```python
import matplotlib.pyplot as plt
import numpy as np
import vibration_toolbox as vtb
import scipy.io as sio
```

```python
%matplotlib inline
mat_contents=sio.loadmat('Case2.mat')
a = mat_contents['Time_domain']
b = mat_contents['Time_chan_2']
plt.plot(a,b)
plt.grid('on')
plt.ylabel('$b$')
plt.xlabel('$a$')
plt.title('$b$ versus $a$')
plt.show()
```
